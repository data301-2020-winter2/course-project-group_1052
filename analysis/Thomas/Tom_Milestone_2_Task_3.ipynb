## What do we want to change in our dataframe?
1. Remove kbLyric, ogLyric, and songName
2. rename badword to badWord
3. sort by year
4. subsort by artist name alphabetically

```python
df = (
    kidzbop_cens.drop(columns=['songName', 'ogLyric', 'kbLyric'])
      .rename(columns={'badword':'badWord'})
      .sort_values("ogArtist", ascending = True)
      .sort_values("year", ascending = True)
     )
df.head(10)
```

```python
def load_and_process(path):
    df1 = pd.read_csv(path)
    
    df2 = (
        df1.drop(columns=['songName', 'ogLyric', 'kbLyric'])
            .rename(columns={'badword':'badWord'})
            .sort_values("ogArtist", ascending = True)
            .sort_values("year", ascending = True)
    )
    return df2
```

```python
load_and_process('data/raw/KB_censored-lyrics.csv')
```
