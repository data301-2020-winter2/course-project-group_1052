## What to change for df

1. Drop isPresent rows: they're redundant and only equal to one value
2. Rename badword to badWord - consistent column naming conventions
3. Sort by values year
4. Show only certain columns?
5. Show 2012+

```python
df = (kidzbop_censored.drop(columns='isPresent')
      .rename(columns={'badword':'badWord'})
      .sort_values(by='year', ascending = False)
      .loc[lambda x: x['year']>=2012]
      .reset_index(drop = True)
      .loc[:,'ogArtist':'year']
     )
df.head()
```

```python
def load_and_process(path):
    df1 = pd.read_csv(path)
    
    df2 = (df1.drop(columns='isPresent')
          .rename(columns={'badword':'badWord'})
          .sort_values(by='year',ascending = False)
          .loc[lambda x: x['year'] >= 2012]
          .reset_index(drop=True)
          .loc[:,'ogArtist':'year']
          )
    return df2
```

```python
load_and_process('data/raw/KB_censored-lyrics.csv')
```
